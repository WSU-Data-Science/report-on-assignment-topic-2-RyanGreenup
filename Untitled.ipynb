Inverse Gaussian:

$$
\left(\frac{y-\mu}{\sigma}\right)^{2}=-2\ln\left(\sigma\sqrt{2\pi}x\right)
$$

Consider two similar but different distributions:

````julia
log(2.718)
````

````julia
function gauss(x; μ=0, σ=1)
    return 1/(σ * sqrt(2*π)) * exp(-0.5 * ((x-μ)/σ)^2)
end

function sigmoid(x; σ=1)
    return 1/(1+exp(-x*σ))
end

function sigmoid_d(x)
    return (exp(-x))/(1+exp(-x))^2
end

function inv_sig(x)
    return log(x/(1-x))
end
````

````julia
# Plot the Sigmoid Curve ...................................
xv = [range(-5, stop=5, length=1000)...]
yv = [sigmoid(x) for x in xv]
ydv = diff(yv)

# Plot the Sigmoid
p1 = Plots.plot(xv,yv, legend = false, title = "Sigmoid Curve")

# Plot the derivative
deleteat!(xv, 1)
p2 = Plots.plot(xv,ydv, legend = false, title = "Sigmoid Derivative")


# Plot the Gaussian Curve ..................................

xv = [range(-5, stop=5, length=1000)...]
yv = [gauss(x) for x in xv]
yiv = cumsum(yv)

# Plot the Gaussian PDF
p3 = Plots.plot(xv,yv, legend = false, title = "Gaussian Probability Density")

# Plot the Integral
# deleteat!(xv, 1)
p4 = Plots.plot(xv,yiv, legend = false, title = "Gaussian Distribution")


# Perform the plotting
Plots.plot(p1,p2, p3, p4, layout = (2,2))

````

The Sigmoid curve is often used for modelling population [@frankgiordanoFirstCourseMathematical2014, Ch. 11], but let's assume that we had some cause to use it to model a statistical distribution instead, in order to do so, we could find the inverse function and provide it with uniform input values $\left[0,1\right]$:

$$
F\left(x\right)=\frac{1}{1+e^{-x}}\implies F^{-1}\left(x\right)=\ln\left(\frac{x}{1-x}\right)
$$

````julia


unif_vals = rand(Uniform(0,1), 100)
sig_vals = [inv_sig(x) for x in unif_vals]

p1 = Plots.histogram(unif_vals, title = "Uniform Values", legend = false, normalize = true);
p2 = Plots.histogram(sig_vals, title = "Sigmoid Distribution", legend = false, normalize = true);

Plots.plot(p1, p2, layout = (2,1), title = "Uniform and Sigmoid Distribution", normalize = true)
````

````julia


unif_vals = rand(Uniform(0,1), 100)
sig_vals = [x->((2*sqrt(2-x))/(sqrt(x))) for x in unif_vals]
,
p1 = Plots.histogram(unif_vals, title = "Uniform Values", legend = false, normalize = true);
p2 = Plots.histogram(sig_vals, title = "Sigmoid Distribution", legend = false, normalize = true);

Plots.plot(p1, p2, layout = (2,1), title = "Uniform and Sigmoid Distribution", normalize = true)
````

This is one strategy to create a distribution of random values from a uniform distribution, note however that this distribution is **not** normal, it follows a sigmoid curve:

````julia
vals = rand(Normal(0, 1), 10^5);
Plots.histogram(vals, normalize = true, color = "light blue",
                bins = 30, title = "Comparison of Gaussian and Sigmoid Curve",
                label = "Gaussian Data")


# Plot a bell curve
xv = [range(-5, stop=5, length=1000)...]
yv = [gauss(x) for x in xv]
yv_d = [exp(-x)/(1+exp(-x))^2 for x in xv]
Plots.plot!(xv,yv, lw = 6, color = "indianred", label = "Gaussian PDF")

# Plot a sigmoid
yv = [sigmoid_d(x) for x in xv]
Plots.plot!(xv,yv, lw = 6, color = "purple", label = "∂/∂x-Sigmoid ")

````

## ODE's

````julia
using SymPy
x = Sym("x")
````

````julia
# Import the Package
using SymPy

# Create the Functions
f, g = symbols("f g", cls=sympy.Function)
# Create the Constants
k, M = symbols("k, M", cls = sympy.Symbol)
# Make the variables
@vars x

# State the Equation
diffeq = Eq(f(x).diff(x),k*f(x)*(M-f(x)))
# diffeq        

# Solve the ODE
dsolve(diffeq, f(x))
````
